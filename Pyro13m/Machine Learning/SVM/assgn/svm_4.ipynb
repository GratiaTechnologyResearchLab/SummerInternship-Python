```python
import h5py
import numpy as np
# load h5py file

def load_h5py(filename):
    with h5py.File(filename, 'r') as hf:
        global X
        global Y
        X= hf['x'][:]
        Y= hf['y'][:]
        return X,Y

load_h5py('data_4.h5')
    
    
#FIT THE CLASSIFIER AND TUNE IT USING 'gamma' FOR THE BEST FIT.(YOU CAN USE OTHER ATTRIBUTES TOO)




from sklearn.svm import SVC
classifier=SVC(kernel = 'rbf',gamma=100,random_state=0)
classifier.fit(X,Y)



import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

x_set,y_set=X,Y

x1,x2=np.meshgrid(np.arange(start=x_set[:].min()-1,stop=x_set[:].max()+1,step=0.01),np.arange(start=x_set[:].min()-1,stop=x_set[:].max()+1,step=0.01))


plt.contourf(x1,x2,classifier.predict(np.array([x1.ravel(),x2.ravel()]).T).reshape(x1.shape),alpha=0.80,cmap=ListedColormap(('red','blue')))


plt.xlim(x1.min(),x1.max())
plt.ylim(x2.min(),x2.max())

#To plot the points

#for i,j in enumerate(np.unique(y_set)):
    #plt.scatter(x_set[y_set==j,0],x_set[y_set==j,1],c=ListedColormap(('white','yellow'))(i),label=j)

plt.title('***SVM MODEL-4***')
plt.xlabel('x-parameter')
plt.ylabel('y-parameter')
plt.show()



plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```
